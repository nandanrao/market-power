```r
library(mixtools)
library(ggplot2)
library(dplyr)
library(haven)
library(purrr)
library(tidyr)
```

```r
d <- read_dta('./dat/08/set_d.dta') %>% select(earn, ind, tempall, hours_m)
```

```r
l <- d %>%
#     filter(hours_m > 80) %>%
    filter(!is.na(earn)) %>%
    mutate(earn = zap_labels(earn)) %>%
    group_by(tempall, ind) %>% 
    filter(n() > 5000 & !is.na(ind) & !is.na(tempall)) %>%
    nest() 
```

```r
m <- map(l$data, function (df) normalmixEM(df$earn))
```

```r
plots <- map(m, function (x) ggplot(data.frame(x = x$x), aes(x=x)) + 
             geom_density() + 
             stat_function(colour = "red", fun=dnorm, n=101, args = list(mean= x$mu[1], sd = x$sigma[1])) + 
             stat_function(colour = "blue", fun=dnorm, n=101, args = list(mean= x$mu[2], sd = x$sigma[2]))
            )                             
```
