```r
get.mixes <- function (d) {
    l <- d %>%
        na.omit() %>%
        filter(hours_m > 80) %>%
        filter(ind %in% groups) %>%
        filter(!is.na(earn)) %>%
        mutate(earn = zap_labels(earn)) %>%
        group_by(ind) %>% 
        nest()
    map(l$data, function (df) normalmixEM(df$earn))
}
```

```r
stats.90 <- map_df(get.mixes(d.1990), function (x) list(mu = x$mu, sigma = x$sigma))
```

```r
stats.08 <- map_df(get.mixes(d.2008), function (x) list(mu = x$mu, sigma = x$sigma))
```
